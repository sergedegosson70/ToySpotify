```python
df = pd.read_csv('c:/users/k_sego/titanic.csv', sep=";")
df.head()
```

```python
X= df #.drop('survived', axis=1)
y = df['survived']
X.head()
```

```python

y = pd.DataFrame(y)
#y['survived'] = y['survived'].astype(str)
y.info()
```

```python
categorical_features = df.select_dtypes(include='object')
categorical_feature_names = list(categorical_features.columns)
categorical_feature_names
```

```python

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))])

numeric_features = df.drop(categorical_feature_names, axis=1)
numeric_feature_names = list(numeric_features.columns)
numeric_feature_names
```

```python
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())])
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_feature_names),
        ('cat', categorical_transformer, categorical_feature_names)])
```

```python
clf = Pipeline(steps=[('preprocessor', preprocessor),
                      ('classifier', LinearRegression())])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python

#X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0)
```

```python

clf.fit(X_train, y_train)
print("model score: %.3f" % clf.score(X_test, y_test))
```
