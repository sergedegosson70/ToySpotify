```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:/users/k_sego/hour.csv", sep=";")
```

```python
df5 = pd.DataFrame(columns=['hour_display'])

for x in df['hour']:
    x = str(x) + ":00 - " + str(x) + ":59"
    print(x) #for test
    df5.append([{'hour_display': x}], ignore_index=True )

print(df5) #for test
df.append(df5)
```

```python
def hour_disply(x):
    return str(x) + ":00 - " + str(x) + ":59"

df['hour_display'] = df['hour'].apply(lambda x: hour_disply(x))
```
