## Statistics by DepartureLineTransportMode

````python
AllTransportModes_null   = AllTransportModes.BoardingsAPC.isnull().groupby([AllTransportModes['ContractName'],AllTransportModes['DepartureLineTransportMode'],AllTransportModes['DayGroupNumber'],AllTransportModes['Year'],AllTransportModes['Month']]).sum().astype(int).reset_index(name='count_nulls')
AllTransportModes_rows   = AllTransportModes.groupby(['ContractName', 'DepartureLineTransportMode','DayGroupNumber','Year','Month']).size().reset_index(name='count_rows')
AllTransportModes_counts = AllTransportModes_null.merge(AllTransportModes_rows, left_on=['ContractName','DepartureLineTransportMode', 'DayGroupNumber','Year','Month'], right_on=['ContractName','DepartureLineTransportMode', 'DayGroupNumber','Year','Month'], how = 'inner')
AllTransportModes_counts['Percentage Null (%) (Non-imputated)'] = round((AllTransportModes_counts['count_nulls']/AllTransportModes_counts['count_rows'])*100,1)
````

````python
AllTransportModes_counts = AllTransportModes_counts.drop(['count_nulls','count_rows'],axis=1)
````

````python
AllTransportModes_null_imp  = AllTransportModes.Boardings.isnull().groupby([AllTransportModes['ContractName'],AllTransportModes['DepartureLineTransportMode'],AllTransportModes['DayGroupNumber'],AllTransportModes['Year'],AllTransportModes['Month']]).sum().astype(int).reset_index(name='count_nulls')
AllTransportModes_rows_imp = AllTransportModes.groupby(['ContractName','DepartureLineTransportMode', 'DayGroupNumber','Year','Month']).size().reset_index(name='count_rows')
AllTransportModes_counts_imp = AllTransportModes_null_imp.merge(AllTransportModes_rows_imp, left_on=['ContractName','DepartureLineTransportMode', 'DayGroupNumber','Year','Month'], right_on=['ContractName','DepartureLineTransportMode', 'DayGroupNumber','Year','Month'], how = 'inner')
AllTransportModes_counts_imp['Percentage Null (%) (Imputated)'] = round((AllTransportModes_counts_imp['count_nulls']/AllTransportModes_counts_imp['count_rows'])*100,1)
d = {'Origin':{'both':'Both','left_only':'Tram_imp','right_only':'Tram'}}

AllTransportModes_statistics =AllTransportModes_counts_imp.merge(AllTransportModes_counts, on = ['ContractName','DepartureLineTransportMode', 'DayGroupNumber','Year','Month'], how ="inner", indicator = True).rename(columns={'_merge':'Origin'}).replace(d)
AllTransportModes_statistics = AllTransportModes_statistics.rename(columns ={'count_nulls_x':'count_nulls_APC2','count_rows_x':'count_rows_APC2','count_nulls_y':'count_nulls_ORG','count_rows_y':'count_null_ORG'})
````

````python
import dash
import dash_core_components as dcc
import dash_html_components as html
import plotly.graph_objs as go
````

````python
Avtal_options    = AllTransportModes_statistics_Tram["ContractName"].unique()
Year_options     = AllTransportModes_statistics_Tram["Year"].astype(str).unique()
Dagtyp_options   = AllTransportModes_statistics_Tram["DayGroupNumber"].astype(str).unique()
````

````python
app = dash.Dash()

app.layout = html.Div([
    html.H2("Saknade data för Lokalbanorna i FactTMSPlannedCarDepartureAP2"),
    html.Div(
        [
            dcc.Dropdown(
                id="ContractName",
                options=[{
                    'label': i,
                    'value': i
                } for i in Avtal_options],
                value='All ContractNames'
            ),
            dcc.Dropdown(
                id="Year",
                options=[{
                    'label': i,
                    'value': i
                } for i in Year_options],
                value='All Years'
             ),
            dcc.Dropdown(
                id="DayGroupNumber",
                options=[{
                    'label': i,
                    'value': i
                } for i in Dagtyp_options],
                value='All DayGroupNumbers'
            ),
        ],
        style={'width': '25%',
               'display': 'inline-block'}),
     dcc.Graph(id='funnel-graph'),
    ])

@app.callback(
    dash.dependencies.Output('funnel-graph', 'figure'),
    [dash.dependencies.Input('ContractName','value'),
    dash.dependencies.Input('Year','value'),
    dash.dependencies.Input('DayGroupNumber','value')])
    
def update_graph(ContractName,Year,DayGroupNumber):
    if ((ContractName == "All ContractNames") & (Year == "All Years") & (DayGroupNumber == "All DayGroupNumbers")):
        AllTransportModes_statistics_Tram_plot = AllTransportModes_statistics_Tram.copy()
    
    elif ((ContractName == "All ContractNames") & (Year == "All Years")):
        AllTransportModes_statistics_Tram_plot = AllTransportModes_statistics_Tram[(AllTransportModes_statistics_Tram['DayGroupNumber'].astype(str)==DayGroupNumber)]
        
    elif ((ContractName == "All ContractNames") & (DayGroupNumber == "All DayGroupNumbers")):
        AllTransportModes_statistics_Tram_plot = AllTransportModes_statistics_Tram[(AllTransportModes_statistics_Tram['Year'].astype(str)==Year)]
    
    elif ((Year == "All Years") & (DayGroupNumber == "All DayGroupNumbers")):
        AllTransportModes_statistics_Tram_plot = AllTransportModes_statistics_Tram[(AllTransportModes_statistics_Tram['ContractName']==ContractName)]
                             
    elif ((ContractName == "All ContractNames")):
        AllTransportModes_statistics_Tram_plot = AllTransportModes_statistics_Tram[(AllTransportModes_statistics_Tram['Year'].astype(str)==Year) & (AllTransportModes_statistics_Tram['DayGroupNumber'].astype(str)==DayGroupNumber)]
    
    elif ((Year == "All Years")):
        AllTransportModes_statistics_Tram_plot = AllTransportModes_statistics_Tram[(AllTransportModes_statistics_Tram['ContractName']==ContractName) & (AllTransportModes_statistics_Tram['DayGroupNumber'].astype(str)==DayGroupNumber)]
                                            
    elif ((DayGroupNumber == "All DayGroupNumbers")):
        AllTransportModes_statistics_Tram_plot = AllTransportModes_statistics_Tram[(AllTransportModes_statistics_Tram['ContractName']==ContractName) & (AllTransportModes_statistics_Tram['Year'].astype(str)==Year)]
    
    elif ((ContractName != "All ContractNames") & (Year != "All Years") & (DayGroupNumber != "All DayGroupNumbers")):
        AllTransportModes_statistics_Tram_plot = AllTransportModes_statistics_Tram[(AllTransportModes_statistics_Tram['ContractName']==ContractName) & (AllTransportModes_statistics_Tram['Year'].astype(str)==Year) & (AllTransportModes_statistics_Tram['DayGroupNumber'].astype(str)==DayGroupNumber)]
        
    trace1 = go.Bar(x=AllTransportModes_statistics_Tram_plot['Month'], y=AllTransportModes_statistics_Tram_plot['Percentage Null (%) (Imputated)'], name='Percentage Null (%) (Imputated)',marker_color='red')
    trace2 = go.Bar(x=AllTransportModes_statistics_Tram_plot['Month'], y=AllTransportModes_statistics_Tram_plot['Percentage Null (%) (Non-imputated)'], name='Percentage Null (%) (Non-imputated)',marker_color='green')

    return {
        'data': [trace1, trace2],
        'layout':
        go.Layout(
            title='Jämförelse saknad data {} år {} och dagtyp {}'.format(ContractName, Year, DayGroupNumber),
            barmode ='group')
    }


if __name__ == '__main__':
    app.run_server()
````

````python
BUS_query = """SELECT 
      f.[CalendarDateKey2]
      ,[ContractName]
      ,[DepartureLineTransportMode]
      ,[DepartureLineNumber]
      ,f.[SLDayTypeCode]
      ,[DayGroupNumber]
      ,[DepartureTimetablePeriodCode]
      ,[HeadTimeTablePeriodCode]
      ,[HasObservations]
      ,[HasAllStopsAPC]
      ,[RatioAPCStops]
      ,[IsDepartureAPC]
      ,[CoveredDistanceObs]
      ,[SeatingCapacityObs]
      ,[MaxCapacityObs]
      ,[BoardingsAPC]
      ,[PersonMeterAPC]
      ,[Boardings]
      ,[PersonMeter]
      ,cd.MonthOfYear as Month
  FROM [RUST2].[dbo].[FactTMSPlannedCarDepartureAPC2] as f
inner join DimCalendarDate cd on f.CalendarDateKey = cd.CalendarDateKey
where f.DepartureLineTransportMode = 'BUS'
and f.ContractName  IN ('Bromma','Ekerö','Handen','Huddinge/Botkyrka/Söderort')
and f.[CalendarDateKey2]>'20191231'
"""
        
        
BUS_Group1_query = pd.read_sql(BUS_query, sql_conn)
````
